```python
def runs_scored(batsman_name):
  return data.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1).index
```

```python
runs_scored('MS Dhoni')
```

```python
death_over = data[data['over']>15]
```

```python
batsman_all = death_over.groupby('batsman')['batsman_runs'].count()
```

```python
x = batsman_all > 200
top_death_batsman = batsman_all[x].index.tolist()
```

```python
 final = data[data['batsman'].isin(top_death_batsman)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
runs
```

```python
balls = final.groupby('batsman')['batsman'].count()
balls
```

```python
strike_rate = runs*100/balls
strike_rate.sort_values(ascending=False)
```

```python
six = sixes.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count').sort_values(by='over',ascending=False)
```
