#### Columnwise Analysis - Chekcing missing values and data types

```python
def analyse_column(names):
  col_analysis = []
  for col in names:
    col_data = df[col]
    col_type = col_data.dtype
    unique_values = col_data.nunique()
    miss_values = col_data.isnull().sum()

    col_analysis.append({
        "Column Name": col,
        "Data Type": str(col_type),
        "Unique Values": unique_values,
        "Missing Values": miss_values
    })
  column_analysis = pd.DataFrame(col_analysis)
  return column_analysis
```

```python
#Checking the unique values and missing values - for categorical columns
cat_col_analysis = analyse_column(cat_col)
cat_col_analysis
```

```python
#Checking the unique values and missing values - for numerical columns
num_col_analysis = analyse_column(num_col)
num_col_analysis
```

```python
#Handling missing values
df.drop('CAMPAIGN_NBR',inplace = True,axis = 1)
```

```python
for i in range(0,100):
  if df['TOTALCOST'][i]=='':
    df['TOTALCOST'] = df['REPORTING_COST']
  else:
    df['TOTALCOST'] = df['TOTALCOST']
```

```python
missing = df.isnull().sum()
missing_col = missing[missing>0].index.to_list()
missing_col
```

```python
for col in missing_col:
  df.fillna({col : df[col].mode()[0]},inplace = True)
```

### Vizualizing Distributions

```python
num_col = ['TRANSACTION_ID',
 'DEALER_REGION',
 'REPAIR_AGE',
 'KM',
 'COMPLAINT_CD_CSI',
 'ORD_SELLING_SRC_CD',
 'GLOBAL_LABOR_CODE',
 'REPORTING_COST',
 'TOTALCOST',
 'LBRCOST',
 'TRANSMISSION_SOURCE_PLANT',
 'SRC_TXN_ID',
 'SRC_VER_NBR',
 'TRANSACTION_CNTR',
 'LAST_KNOWN_DELVRY_TYPE_CD',
 'NON_CAUSAL_PART_QTY',
 'SALES_REGION_CODE']
#--- 1. Correlation Heat Map ---
plt.figure(figsize=(12, 8))
sns.heatmap(df[num_col].corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
# --- 2. Distribution of TOTALCOST ---
plt.figure(figsize=(8, 5))
sns.histplot(df["TOTALCOST"].dropna(), bins=20, kde=True, color="blue")
plt.title("Distribution of TOTALCOST")
plt.xlabel("TOTALCOST")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.set(style="whitegrid")
# --- 3. Sales by Region ---
plt.figure(figsize=(8, 5))
sns.countplot(x=df['SALES_REGION_CODE'], palette='viridis')
plt.title('Sales by Region')
plt.xlabel('Region Code')
plt.ylabel('Number of Sales')
plt.show()
```

```python
# --- 4 . Boxplot of TOTALCOST by PLATFORM ---
plt.figure(figsize=(10, 5))
sns.boxplot(x="PLATFORM", y="TOTALCOST", data=df, palette="Set2")
plt.xticks(rotation=45)
plt.title("TOTALCOST by PLATFORM")
plt.show()
```

```python
# --- 5. Most Common Repairs ---
sns.countplot(y=df['GLOBAL_LABOR_CODE_DESCRIPTION'], order=df['GLOBAL_LABOR_CODE_DESCRIPTION'].value_counts().index)
plt.title('Most Common Repairs')
plt.xlabel('Count')
plt.ylabel('Repair Description')
plt.show()
```

```python
# --- 6. Top 5 Selling Vehicle Models ---
top_models = df['VIN_MODL_DESGTR'].value_counts().head(5)
plt.figure(figsize=(8, 5))
sns.barplot(x=top_models.index, y=top_models.values, palette='viridis')
plt.title('Top 5 Selling Vehicle Models')
plt.xlabel('Vehicle Model')
plt.ylabel('Number of Sales')
plt.show()
```

```python
# --- 7. Mileage vs. Repair Age ---
sns.scatterplot(x=df['KM'], y=df['REPAIR_AGE'])
plt.title('Mileage vs. Repair Age')
plt.xlabel('Mileage')
plt.ylabel('Repair Age')
plt.show()
```

```python
# --- 8. Cost Over Time ---
plt.figure(figsize=(10, 5))
sns.lineplot(x=df['REPAIR_DATE'], y=df['TOTALCOST'], marker='o', color='b')
plt.title('Cost Over Timee')
plt.xlabel('Date')
plt.ylabel('Total Cost')
plt.xticks(rotation=45)
plt.show()
```
